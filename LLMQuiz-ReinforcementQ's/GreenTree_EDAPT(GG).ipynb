This notebook explores the Titanic dataset by performing common EDA steps including data inspection, visualizing distributions, and understanding feature relationships

*Loading Libraries and Dataset*



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Titanic dataset
df = pd.read_csv("Titanic-Dataset.csv")
```

*Displaying the structure of the dataset*

```python
df.info()

# Summary of data
df.describe()
```

*Age distribution Visualisation*

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="Age", kde=True, bins=30, color='green')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

*Visualisation of Gender vs Survival*

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Sex", hue="Survived")
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

*Visualisation of Numeric features*

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
